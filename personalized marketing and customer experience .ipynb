```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import random
import sys
import matplotlib.pyplot as plt  # Import matplotlib for plotting

# 1. Data Representation (Illustrative)
# Sample Customer Data (Extracted/Simulated)
customers_data = {
    'customer_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily', 'Frank', 'Grace'],
    'purchase_history': [['Product A', 'Product C'], ['Product B'],
                       ['Product A', 'Product D', 'Product E'], [],
                       ['Product C', 'Product B'], ['Product A', 'Product F'],
                       ['Product G']],
    'browsing_history': [['Category X', 'Category Y'], ['Category Y'],
                        ['Category X'], ['Category Z'],
                        ['Category Y', 'Category Z'], ['Category X'],
                        ['Category Y']],
    'preferences': [{'marketing_consent': True, 'language': 'en'},
                    {'marketing_consent': False, 'language': 'en'},
                    {'marketing_consent': True, 'language': 'es'},
                    {'marketing_consent': True, 'language': 'en'},
                    {'marketing_consent': True, 'language': 'en'},
                    {'marketing_consent': True, 'language': 'en'},
                    {'marketing_consent': False, 'language': 'en'}]
}
customers = pd.DataFrame(customers_data)

# Sample Product Catalog (Extracted/Simulated)
products_data = {
    'product_id': ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
    'name': ['Product A', 'Product B', 'Product C', 'Product D',
             'Product E', 'Product F', 'Product G'],
    'category': ['Category X', 'Category Y', 'Category X', 'Category Z',
                 'Category Z', 'Category Y', 'Category X'],
    'description': ['Description of Product A', 'Description of Product B',
                    'Description of Product C', 'Description of Product D',
                    'Description of Product E', 'Description of Product F',
                    'Description of Product G']
}
products = pd.DataFrame(products_data)


# 2. Personalization Engine
def get_content_based_recommendations(customer_id, products, customers):
    customer = customers[customers['customer_id'] == customer_id].iloc[0]
    recommended_products = []
    if customer['purchase_history']:
        purchased_categories = list(
            set(products[products['name'].isin(customer['purchase_history'])]['category']))
        recommended_products = products[
            (products['category'].isin(purchased_categories)) & (
                ~products['name'].isin(customer['purchase_history']))]['name'].tolist()
    else:
        recommended_products.append("No purchase history. Showing popular products.")
        recommended_products.extend(products['name'].tolist())
    return recommended_products


def personalize_offers(customer_id, customers):
    customer = customers[customers['customer_id'] == customer_id].iloc[0]
    offers = []
    if customer['browsing_history']:
        last_browsed_category = customer['browsing_history'][-1]
        offers.append(f"Special offer on: {last_browsed_category} products!")
    if customer['preferences']['marketing_consent']:
        offers.append("Get 10% off your next purchase!")
    return offers


# 3. Chatbot Interaction
def chatbot_response(user_input, knowledge_base, customer_id=None):
    user_input_lower = user_input.lower()
    for key, response_func in knowledge_base.items():
        if key in user_input_lower:
            return response_func(user_input, customer_id)
    return "I'm sorry, I don't have information on that. How else can I help?"


# Chatbot Knowledge Base
def handle_product_inquiry(user_input, customer_id):
    return "I can help you find products."


def handle_account_inquiry(user_input, customer_id):
    return "I can help you with your account."


def handle_recommendation_request(user_input, customer_id):
    return "I can give you recommendations."


chatbot_knowledge_base = {
    "product": handle_product_inquiry,
    "account": handle_account_inquiry,
    "recommendation": handle_recommendation_request,
    "help": lambda user_input, customer_id: "I am here to assist you."
}


# 4. Privacy
def check_marketing_consent(customer_id, customers):
    customer = customers[customers['customer_id'] == customer_id].iloc[0]
    return customer['preferences']['marketing_consent']


# 5. Enhanced Functions (Phase 4)
# AI Model Enhancement
def retrain_ai_model(customers, products):
    new_products = pd.DataFrame({
        'product_id': ['H', 'I'],
        'name': ['Product H', 'Product I'],
        'category': ['Category Z', 'Category X'],
        'description': ['Description of Product H',
                        'Description of Product I']
    })

    new_customers = pd.DataFrame({
        'customer_id': [108, 109],
        'name': ['Harry', 'Ivy'],
        'purchase_history': [['Product B', 'Product H'], ['Product I']],
        'browsing_history': [['Category Z'], ['Category X']],
        'preferences': [{'marketing_consent': True, 'language': 'en'},
                        {'marketing_consent': False, 'language': 'en'}]
    })

    updated_products = pd.concat([products, new_products], ignore_index=True)
    updated_customers = pd.concat([customers, new_customers], ignore_index=True)

    print("AI Model Retrained with new data.")
    return updated_customers, updated_products


def optimize_ai_model():
    print("AI Model Optimized for speed and accuracy.")


# Chatbot Optimization
def improve_chatbot_response_time():
    print("Chatbot response time improved.")


def enhance_nlp_capabilities():
    global chatbot_knowledge_base
    chatbot_knowledge_base["help"] = \
        lambda user_input, customer_id: "I am here to assist you."
    print("Chatbot NLP capabilities enhanced.")


# Omnichannel
def synchronize_data_across_channels():
    print("User data synchronized across web, mobile, email, and social media.")


def optimize_api_calls():
    print("API calls optimized for smoother and faster data retrieval.")


# Security
def implement_robust_encryption():
    print("Robust encryption implemented for data security.")


def perform_security_testing():
    print("Security testing performed to ensure data integrity and privacy.")


# Performance Testing and Metrics
def perform_load_testing():
    print("Load testing performed to handle increased user traffic.")


def collect_performance_metrics():
    metrics = {
        "response_time": 0.1,  # Numerical value for plotting
        "system_stability": 99.99,  # Numerical value
        "error_rate": 0.005,  # Numerical value
        "data_processing_capacity": 1500  # Numerical value
    }
    print("Performance metrics collected:", metrics)
    return metrics


def gather_user_feedback():
    feedback = [
        {"user_id": 101, "rating": 5,
         "comment": "Great recommendations! Faster response!"},
        {"user_id": 102, "rating": 4, "comment": "Chatbot was helpful. Improved NLP."},
        {"user_id": 103, "rating": 3,
         "comment": "Personalization could be better. More variety needed."}
    ]
    print("User feedback gathered:", feedback)
    return feedback


# ---Example Usage ---
print("Enter the customer ID (101 to 109):")
customer_id_input = input()

try:
    customer_id = int(customer_id_input)
    if customer_id not in customers['customer_id'].values:
        print("Invalid customer ID.")
        customer_id = 101
except ValueError:
    print("Invalid input. Using default customer ID (101).")
    customer_id = 101

# Phase 4 Enhancements
updated_customers, updated_products = retrain_ai_model(customers, products)
optimize_ai_model()
improve_chatbot_response_time()
enhance_nlp_capabilities()
synchronize_data_across_channels()
optimize_api_calls()
implement_robust_encryption()
perform_security_testing()
perform_load_testing()
metrics = collect_performance_metrics()
feedback = gather_user_feedback()

print("\n--- Recommendations for Customer", customer_id, "---")
recommendations = get_content_based_recommendations(customer_id,
                                                   updated_products,
                                                   updated_customers)
print(recommendations)

print("\n--- Personalized Offers for Customer", customer_id, "---")
offers = personalize_offers(customer_id, updated_customers)
print(offers)

print("\n--- Chatbot Interaction ---")
user_input = input("Enter your chatbot query: ")
chatbot_reply = chatbot_response(user_input, chatbot_knowledge_base,
                                customer_id)
print(chatbot_reply)

print("\n--- Marketing Consent Check ---")
if check_marketing_consent(customer_id, updated_customers):
    print(f"Customer {customer_id} has consented to marketing.")
else:
    print(f"Customer {customer_id} has NOT consented to marketing.")


# --- Graph Output (Example) ---
# Create a simple bar chart for performance metrics
plt.figure(figsize=(8, 6))
plt.bar(metrics.keys(), metrics.values())
plt.title('System Performance Metrics')
plt.ylabel('Value')
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()

# Create a simple bar chart for user feedback ratings
ratings = [f['rating'] for f in feedback]
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(ratings) + 1), ratings)
plt.title('User Feedback Ratings')
plt.xlabel('User')
plt.ylabel('Rating')
plt.show()
```
